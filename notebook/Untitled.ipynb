```python
import pandas as pd
import seaborn as sns
```

```python
f = '../data/ier_2022.xlsx'
ier = pd.read_excel(f,skiprows=4,skipfooter=7)
cols = ier.columns
nombres = ['Unnamed: 0', 'nombre', 'categoria', 'TIPO \nDE \nCONTRATO',
       'RFC', 'genero', 'sni', 'VIGENCIA - SNI', 'CLAVE\nCVU',
       'EXPEDIENTE\nS N I', 'pride', 'VIGENCIA PRIDE',
       'PRIMER \nCONTRATO\nACADÉMICO', 'antiguedad',
       'ANTIGÜEDAD\nA]CADÉMICA\nDGP', 'FECHA DE NACIMIENTO', 'EDAD',
       'Observaciones']
columnas = dict(zip(columnas,nombres))
ier.rename(columns=columnas,inplace=True)
ier.sni.fillna(-1,inplace=True)
ier.sni.replace('C',0,inplace=True)
ier.sni.replace('EMÉRITO',3,inplace=True)
ier.sni.replace('NO TIENE',-1,inplace=True)
ier.sni.replace('\nSalió a partir de 01/01/17',-1,inplace=True)
ier.sni.replace('\nSalió a partir de 01/01/20',-1,inplace=True)
ier.sni.replace('Salió a partir del 01/01/16',-1,inplace=True)
ier.sni.replace('Salió',-1,inplace=True)
ier.sni.replace('C\n',0,inplace=True)
ier.sni.replace('c',0,inplace=True)

ier.sni = ier.sni.astype({'sni':'int'})
```

```python
sns.catplot(
    data=ier, x="pride", y="sni", hue="genero",
    kind="violin", split=True,
)
```

```python

ax = sns.swarmplot(data=ier, x="antiguedad", y="genero", hue="pride")
```

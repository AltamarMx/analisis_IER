```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS
```

```python
f =  '../data/Seminarios2022.xlsx'
df = pd.read_excel(f)
df.info()
```

```python
sns.catplot(x='Origen Invitadx',kind='count',data=df)
```

```python
text = df['Area de conocimiento'].dropna().values
text = str(text)
text = text.replace(",","").replace("'","").replace('\n','')
wordcloud = WordCloud().generate(str(text))

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
